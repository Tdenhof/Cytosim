# Load Pretrained Model 

```python
import torch
PATH = '/content/drive/Shareddrives/Cytosim/for_machine_learning/training/train/exp3/weights/best.pt'
model = torch.hub.load('ultralytics/yolov5', 'custom', path='/content/gdrive/Shareddrives/Cytosim/for_machine_learning/training/train/exp3/weights/best.pt')  # local model
```

```python
results = model('/content/gdrive/Shareddrives/Cytosim/for_machine_learning/data/images/train/binding_rate=15_t2_annotated_movie0094.png')
```

```python
import glob
import pandas as pd
def predictMovie(model, folder):
  finalResults = []
  imgs = glob.glob(folder + '/*.png')
  for img in imgs: 
    results = model(img)
    df = results.pandas().xyxy[0]
    df['img'] = img
    finalResults.append(df)
  preds = pd.concat(finalResults)
  return preds

```

```python
preds =  predictMovie(model,'/content/gdrive/Shareddrives/Cytosim/for_machine_learning/hole /binding_rate=20_t3')
```
